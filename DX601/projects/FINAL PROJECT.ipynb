### Problem 5 (20 points)

Build an ordinary least squares regression for the target using all the input columns.
Report the mean squared error of the model over the whole data set.
Plot the actual values vs the predicted outputs to compare them. 

```python
#YOUR CODE HERE

from sklearn.linear_model import LinearRegression #import the linear regression model from sklearn

df["species_num"] = df["species"].astype("category").cat.codes #convert the species column to a categorical type
                                                                #then use cat.codes to convert to numeric codes

X = df[["sepal_length", "sepal_width", "petal_length", "petal_width"]] #
y = df["species_num"].values #X is the independent variable, which are the measurements of the iris
                                #y is the dependent variable, which is the species of iris, converted to numeric codes


model = LinearRegression() 
model.fit(X, y) #fit the linear regression model to the data

predict = model.predict(X) #use the fitted model to make predictions on the same data

mean_squared_error = np.mean((y - predict) ** 2) #calculate the mean squared error manually by taking the difference
                                    #between the actual values (y) and the predicted values, 
                                    #squaring the differences, and then taking the mean of differences
print("MSE (manual):", mean_squared_error)

plt.scatter(y, predict)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("OLS Regression: Actual vs Predicted")
plt.show()
```

```python
'''I believe the point of this problem is to see if a set of inputs (the measurements of the iris)
can be used to predict which species of iris it is.

The means squared error is around .04, which is low, but not infallible. With this data, we first needed
to convert the species column to numeric (0, 1, 2) in order to use them in a linear regression model.
The scatter plot of actual vs predicted values shows a great correlation. 

Overall, this linear regression model does a good job of predicting the species of iris based on the measurements'''
```

### Problem 6 (20 points)

Which input column gives the best linear model of the target on its own?
How does that model compare to the model in problem 5?


```python
# YOUR CODE HERE

df["species_num"] = df["species"].astype("category").cat.codes

X_columns = ["sepal_length", "sepal_width", "petal_length", "petal_width"] #list of columns to be used as independent variables
y = df["species_num"].values #y is the dependent variable, which is the species of iris, converted to a number

results = {} #create an empty dictionary to store each MSE


for col in X_columns: #this time, loop each value(measurement) individually
    X_single = df[[col]].values #X_single is the independent variable, which is the single measurement of the iris

    model = LinearRegression() #create model
    model.fit(X_single, y) #predict the model using the single measurement

    pred = model.predict(X_single) #use the fitted model to make predictions on the same data

    mse = np.mean((y - pred) ** 2) #calculate the mean squared error
    results[col] = mse #store the MSE in the dictionary


for k, v in results.items(): #use key, value to print the MSE for each feature
    print(k, "MSE:", v)

best_feature = min(results, key=results.get) #lowest MSE, most useful predictor measurement
print("Best single feature:", best_feature) # show lowest predictor
```

```python
'''So, our results are as follows:
PETAL WIDTH is the smallest error, therefore the best single predictor of iris species
PETAL LENGTH is a close second
SEPAL LENGTH is a distant third, around four times greater the error than petal width
SEPAL WIDTH is the worst predictor, with an error around 10 times greater than petal width in first place

Overall, petal measurements are a far better predictor of iris species the sepal measurements.

Problem 5 shows a full model with all features to predict iris species, and the MSE is around .04,
which is better than any single feature alone, seen in problem 6. This goes to show that using a 
combination of measurements is better than using one alone, and reduces error due to outliers in any
specific region of the data.'''
```

### Problem 8 (40 points)

Perform principal components analysis of the input columns.
Compute how much of the data variation is explained by the first half of the principal components.
Build a linear regression using coordinates computed from the first half of the principal components.
Compare the mean squared error of this model to the previous model.
Plot actual targets vs predictions again. 

This problem depends on material from week 13.

```python
# YOUR CODE HERE

X = df[["sepal_length", "sepal_width", "petal_length", "petal_width"]].values #convert to 2d array
y = df["species_num"].values #already numeric, also ensures a 2d array


pca = PCA() #PCA decomposition, or reveals redundancy and compressible data FIND VARIANCE IN DATA
X_pca = pca.fit_transform(X)


k = X.shape[1] // 2 #2 because that's half of 4 features
explained_variance = np.sum(pca.explained_variance_ratio_[:k])
print("First half Principal Components):", explained_variance)

X_reduced = X_pca[:, :k] #take all 150 rows up to but not including k (2), so columns 0 and 1

model = LinearRegression()
model.fit(X_reduced, y) #predict with model as previous problems

predict = model.predict(X_reduced)


mse_pca = np.mean((y - predict) ** 2) #find mean squared error
print("PCA regression MSE:", mse_pca)

plt.scatter(y, predict)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("PCA Regression Results")
plt.show()
```

```python
'''In this case, we remove the redundancy and show how much information remains after 
reducing the dataset. Or, highly correlated data such as petal length vs width can be reduced 
into one (principal) component.'''
```

### Problem 10 (30 points)

Identify an outlier row in the data set.
You may use any criteria discussed in this module, and you must explain the criteria and how it led to picking this row.
Give a visualization showing how much this row sticks out compared to the other data based on your criteria.

```python
# YOUR CODE HERE

from scipy.stats import zscore

X = df[["sepal_length", "sepal_width", "petal_length", "petal_width"]] #z score shows how far a value 
                                                                        #is from the mean in standard deviations

z = np.abs(zscore(X)) #compute zscore for each feature, convert to absolute value
                        #in stdevs

outlier_index = np.argmax(np.sum(z, axis=1)) #sum z scores for all rows across the four measurements
                                            #gives a score for each
                                            #find row with maximum value (argmax)
                                            #axis=1 is row by row across, separately

print("Outlier Row:")
print(df.iloc[outlier_index]) #print outlier row, iloc=by index position

plt.scatter(df["petal_length"], df["petal_width"])
plt.scatter(df.iloc[outlier_index]["petal_length"],
            df.iloc[outlier_index]["petal_width"], #scatterplot of petal measurements and show
                                                    #outlier in RED
            color="red")                            

plt.xlabel("petal_length")
plt.ylabel("petal_width")
plt.title("Outlier Detection") #labels for scatterplot
plt.show()
```

```python
'''So, this code is finding an outlier by first standardizing z-scores, or how far away (in standard deviations)
each is from the mean. The absolute values are then summed across the four measurements for each flower
(sample)(axis=1 does this in the code). In other words, each flower gets a score based on the four 
featured measurements. 

From there, it is easy to see which flower is most abnormal because it has the highest z-score, or
it is furthest away from normal when normalized in a uniform way with all the other flowers (data).

In other words, the red dot is the data that is least likely to help generalize normals or predictions
for other unseen flowers (or any other data).'''
```
