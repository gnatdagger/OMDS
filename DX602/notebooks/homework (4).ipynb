**Concept:** *Functions with lists and indexing.*

**Task:**

Write a function middlest_element(values) that returns the middle element of a list.

Rules:

* If the list is empty, return `None`.

* If the list length is odd, return the element at the exact middle index.

* If the list length is even, return the element just left of center (example: for length 6, use index 2).

* Then run the provided test cases and print results exactly as shown.

**Sample Output:**

```
middlest_element([1,2,3,4,5]) =  3
middlest_element([1,2,3,4,5,6]) =  3
middlest_element([]) =  None
```


````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def middlest_element(values):
    '''middle element is middle if odd, to the right of middle if even'''
    length = len(values)

    if length == 0:
        return None 
    if length % 2 == 1: #odd
        middle_element = length // 2 #finding center
    else:  #even
        middle_element = (length //2) - 1 

    return values[middle_element]
````

**Problem 7**

**Concept:** *Accumulator pattern with numeric result, list iteration.*

**Task:**

Write a function `variance(values)` that returns the population variance of a list of numbers using this formula:

$$
\text{variance} = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2
$$

Rules:

* Must use loops + accumulator (no `sum()`).

* No imports.

* Must call your `mean(values)` function to get the mean.

* Then run the provided test case and print in the same format.

**Sample Output:**

```
values: [6, 3, 7, 9, 5]
mean: 6.0
variance: 4.0
```


````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def mean(values):
    '''return the mean of a list of numbers'''
    total = 0 
    for v in values: 
        total = total + v 
    return total / len(values) 

def variance(values): #define the given function
  '''return the population variance of a list of numbers'''
  m = mean(values)    #function from above being given a variable
  square_diff = 0 #this is the accumulator value
  for v in values: #same variable as the first function
     square_diff = square_diff + (v - m) ** 2
  return square_diff / len(values) # remember to watch that indentations aren't included in the loop by indenting
````

**Problem 8**

**Concept:** *Accumulator pattern with numeric result, list iteration, index-based loop.*

**Task:**

Write a function `covariance(xs, ys)` that computes the population covariance between two series:

$$
\text{covariance} = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})
$$

Rules:

* If `xs` and `ys` are not the same length, print:
    
    `xs and ys must be the same length`
    
    and return `0`.

* Must use an index-based loop.

* Must use your `mean()` function.

* Then run the given test case and print exactly in the same format.

**Sample Output:**

```
x values: [6, 3, 7, 9, 5]
y values: [4, 1, 7, 5, 0]
covariance: 3.8
```


````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def mean(values): # mean function reused
    '''return the mean of a list of numbers'''
    total = 0 
    for v in values: 
        total = total + v 
    return total / len(values) 

def covariance(xs, ys): #define function given
    '''return the covariance of two list that have equal lengths'''
    if len(xs) != len(ys):
        print("xs and ys must be the same length")
        return 0 # account for uneven list lengths as directed
        
    m_x = mean(xs) #define mean of first list
    m_y = mean(ys) #define mean of second list

    total = 0 #create accumulator variable

    for i in range(len(xs)): #this is an index based loop calling a value within the length of xs
        total = total + (xs[i] - m_x) * (ys[i] - m_y) #accumulator value plus the variance of the indexed value - the mean of each list

    return total / len(xs) #covariance
````

**Problem 12**

**Concept:** *Accumulator pattern to process a string, produce a string result.*

**Task:**

Write a function `cipher(s, n)` that applies a Caesar cipher shift to a plaintext string.

Rules:

* Assume lowercase letters only for shifting (`a` to `z`).

* Do not shift non-letters (spaces stay spaces, punctuation stays punctuation).

* Must wrap around the alphabet if shifting past `z` or before `a`.

* Must use an accumulator string.

* Then run the exact test cases shown and print in the same format.

**Sample Output:**

```
plaintext = abc, ciphered = def
plaintext = abc xyz, ciphered = def abc
```


````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
#string to string
def cipher(s, n): #define given function
    '''apply a shift of n to a lowercase string'''
    result = "" #accumulator for string
    for ch in s: # each character in s
        if 'a' <= ch <= 'z': # only lowercase letters
            shift = (ord(ch) - ord('a') + n) % 26 + ord('a') # had to look this up
            result = result + chr(shift) # conversion
        else:
            result = result + ch
    return result
    
````

**Task:**

Write a function `print_all_decipherings(s)` that prints all 26 possible Caesar shifts of string `s`.

Rules:

* Use a definite loop for shifts `0` through `25`.

* For each shift `i`, compute `ciphered = cipher(s, i)` and print:

    `{i}, {ciphered}`

* Then run these two test cases:

    `s = "def"`

    `s = "oqfcgg hvs ibwjsfgs"`

**Sample Output:**

For `"def"` it must print 26 lines starting with `0, def` and ending with `25, cde`

For `"oqfcgg hvs ibwjsfgs"` it must print 26 lines. One of theses is the name of a Beatles song.



````python
# TO DO: WRITE YOUR CODE/SOLUTION BELOW
def cipher(s, n):
    '''print a shift of n to a lowercase string''' 
    result = "" 
    for ch in s: 
        if 'a' <= ch <= 'z': 
            shift = (ord(ch) - ord('a') + n) % 26 + ord('a')
            result = result + chr(shift) 
        else:
            result = result + ch
    return result #use the function from the last problem


def print_all_decipherings(s): #define given function
    '''print 26 shifts in a string'''
    for i in range(26): #loop shifts from range 0 - 25
        ciphered = cipher(s, i) #altering the function from the last problem
        print(f"{i}, {ciphered}") # print the shift in the loop, then the string that results from the shift
````
